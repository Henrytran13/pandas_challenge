```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
Total = purchase_data["SN"].nunique()
Totals = pd.DataFrame({"Total Players": [Total]})
Totals
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate
Unique_Items = purchase_data["Item ID"].nunique()
average_price = purchase_data["Price"].mean()
num_purchases = purchase_data["Purchase ID"].count()
revenue = purchase_data["Price"].sum()
#DataFrame
New_df= pd.DataFrame({"Number of Unique Items":[Unique_Items], 
                      "Average Price" : [average_price], 
                      "Number of Purchases" : [num_purchases], 
                      "Total Revenue" : [revenue] })
#Clean reformat
New_df["Average Price"]=New_df["Average Price"].map("${:,.2f}".format)
New_df["Total Revenue"]=New_df["Total Revenue"].map("${:,.2f}".format)
New_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Bin for Age Group
age_bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,99.9]
age_labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data["Age Group"]=pd.cut(purchase_data["Age"], age_bins, labels=age_labels)

#Age Group groupby
age_group=purchase_data.groupby("Age Group")
age_group_counts = age_group["SN"].nunique()

#age_group_total = age_group["Age"].sum()
age_percent = age_group_counts/Total*100
age_percent = age_percent.map("{:,.2f}%".format)

Age_group_df=pd.DataFrame({"Total Count":age_group_counts,
                           "Percentage of Players":age_percent})
Age_group_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Age_purchase = purchase_data.groupby("Age Group").nunique()["SN"]
Age = purchase_data.groupby(["Age Group"])

Age_count = Age["Purchase ID"].count()

Age_Total_value = Age["Price"].sum()
#Total_value=Total_value.map("${:,.2f}".format)

Age_Average = Age["Price"].mean()
#Age_Average=Age_Average.map("${:,.2f}".format)

Age_Average_person=Age_Total_value/Age_purchase
#Average_person=Average_person.map("${:,.2f}".format)


Age_df=pd.DataFrame({"Purchase Count":Age_count,
                    "Average Purchase Price":Age_Average,
                    "Total Purchase Value": Age_Total_value,
                     "Avg Total Purchase per Person":Age_Average_person})

Age_df["Average Purchase Price"]=Age_df["Average Purchase Price"].map("${:,.2f}".format)
Age_df["Total Purchase Value"]=Age_df["Total Purchase Value"].map("${:,.2f}".format)
Age_df["Avg Total Purchase per Person"]=Age_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

Age_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Item_data=purchase_data[["Item ID", "Item Name", "Price"]]

Item=Item_data.groupby(["Item ID","Item Name"])


Item_count = Item["Price"].count()

Item_Total_value = Item["Price"].sum()
#Item_Total_value=Item_Total_value.map("${:,.2f}".format)

Item_Average = Item_Total_value/Item_count
#Item_Average = Item_Average.map("${:,.2f}".format)

Item_df=pd.DataFrame({"Purchase Count":Item_count,
                    "Average Purchase Price": Item_Average,
                   "Total Purchase Value": Item_Total_value})

Item_df1=Item_df.sort_values(["Purchase Count"], ascending = False)
Item_df1["Average Purchase Price"]=Item_df1["Average Purchase Price"].map("${:,.2f}".format)
Item_df1["Total Purchase Value"]=Item_df1["Total Purchase Value"].map("${:,.2f}".format)
Item_df1
#Item_df_sort = Item_df.sort_values("Total Purchase Value", ascending=False )
```
