```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()

```

```python
#players_total = purchase_data.count()
#players_total
Total = purchase_data["SN"].nunique()
Totals = pd.DataFrame({"Total": [Total]})
Totals
```

```python
#Calculate
Unique_Items = purchase_data["Item ID"].nunique()
average_price = purchase_data["Price"].mean()
num_purchases = purchase_data["Purchase ID"].count()
revenue = purchase_data["Price"].sum()
#DataFrame
New_df= pd.DataFrame({"Number of Unique Items":[Unique_Items], 
                      "Average Price" : [average_price], 
                      "Number of Purchases" : [num_purchases], 
                      "Total Revenue" : [revenue] })
#Clean reformat
New_df["Average Price"]=New_df["Average Price"].map("${:,.2f}".format)
New_df["Total Revenue"]=New_df["Total Revenue"].map("${:,.2f}".format)
New_df
```

```python
genders_purchase = purchase_data.groupby("Gender").nunique()["SN"]
percent = genders_purchase/Total*100

NewGender_df=pd.DataFrame({"Total count":genders_purchase,
                         "Percent of Player":percent})

NewGender_df["Percent of Player"]=NewGender_df["Percent of Player"].map("{:,.2f}%".format)
#print(genders_purchase)
#print(percent)
NewGender_df
```

```python
#= usa_ufo_df["state_countsstate"].value_counts()
T = purchase_data.groupby(["Gender"])
T_count = T["Purchase ID"].count()

Total_value = T["Price"].sum()
#Total_value=Total_value.map("${:,.2f}".format)

Average = T["Price"].mean()
Average=Average.map("${:,.2f}".format)

Average_person=Total_value/genders_purchase
Average_person=Average_person.map("${:,.2f}".format)

T_count
Total_value
Average
Gender_Purchase_df=pd.DataFrame({"Purchase Count":T_count,
                                 "Average Purchase Price":Average,
                                 "Total Purchase Value": Total_value,
                                 "Avg Total Purchase per Person":Average_person})

Gender_Purchase_df["Total Purchase Value"]=Gender_Purchase_df["Total Purchase Value"].map("${:,.2f}".format)
Gender_Purchase_df
#Average_person
#T_df=pd.DataFrame({"Gender": [T]})
#T_df
```

```python
age_bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,99.9]
age_labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data["Group Age"]=pd.cut(purchase_data["Age"], age_bins, labels=age_labels)
purchase_data 
```

```python
age_group=purchase_data.groupby(["Group Age"])
age_group_counts = age_group["Age"].count()

age_group_total = age_group["Age"].sum()
age_percent = age_group_counts/age_group_total*100
age_percent = age_percent.map("{:,.2f}%".format)

Age_group_df=pd.DataFrame({"Total Count":age_group_counts,
                           "Percentage of Players":age_percent})
Age_group_df
```

```python
Age_purchase = purchase_data.groupby("Group Age").nunique()["SN"]
Age = purchase_data.groupby(["Group Age"])

Age_count = Age["Purchase ID"].count()

Age_Total_value = Age["Price"].sum()
#Total_value=Total_value.map("${:,.2f}".format)

Age_Average = Age["Price"].mean()
Age_Average=Age_Average.map("${:,.2f}".format)

Age_Average_person=Age_Total_value/Age_purchase
#Average_person=Average_person.map("${:,.2f}".format)
Age_purchase
Age_count
#Age_Total_value
#Age_Average
#Age_Average_person
Age_df=pd.DataFrame({"Purchase Count":Age_count,
                    "Average Purchase Price":Age_Average,
                    "Total Purchase Value": Age_Total_value,
                     "Avg Total Purchase per Person":Age_Average_person})

Age_df["Total Purchase Value"]=Age_df["Total Purchase Value"].map("${:,.2f}".format)
Age_df["Avg Total Purchase per Person"]=Age_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
Age_df
```

```python
SN_purchase = purchase_data.groupby("SN").nunique()["Item ID"]
SN = purchase_data.groupby(["SN"])

SN_count = SN["Item ID"].count()
SN_count

SN_Total_value = SN["Price"].sum()
SN_Total_value=SN_Total_value.map("${:,.2f}".format)
#SN_Total_value=SN_Total_value.sort_value(ascending = False)

SN_Average = SN["Price"].mean()
SN_Average = SN_Average.map("${:,.2f}".format)

SN_df=pd.DataFrame({"Purchase Count":SN_count,
                    "Average Purchase Price":SN_Average,
                   "Total Purchase Value": SN_Total_value})
SN_df
#SN_df_sort=SN_df.sort_value("Total Purchase Value", ascending=False )
#SN_df_sort
```
